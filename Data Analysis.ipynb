```python
#Counting LTOT patients
ltot = df.groupby('Patient')
ltot_count=df.groupby('Patient').size()
print("Distribution of LTOT: ", ltot_count)
print("")

# Examine skew of the attribute distributions
skew = df.skew()
print("Skew of attributes:")
print("")
print(skew)
print("")

# Crosstabs
print("Mean values of inputs by pre-diabetes class:")
df2=pd.pivot_table(df,values=['TotalMME', 'AverageDailyMME', 'BenzoFill', 'OpioidFill', 'NonOpioid_Rx', 'NewDoc','CPD_Diagnosis','CHF_Diagnosis','Diabt_Diagnosis','Hyprt_Diagnosis','CAD_Diagnosis','Top5_Diagnosis','AUD_Incidence','Mental_Health_Conditions'],index='Patient',aggfunc=np.mean)
df2.head()
```

```python
# Matrix Plots
from matplotlib import pyplot
names = ['TotalMME', 'AverageDailyMME', 'BenzoFill', 'OpioidFill', 'NonOpioid_Rx', 'NewDoc','CPD_Diagnosis','CHF_Diagnosis','Diabt_Diagnosis','Hyprt_Diagnosis','CAD_Diagnosis','Top5_Diagnosis','AUD_Incidence','Mental Health Conditions']

# Correlation matrix
print("Heatmap of partial correlations between model inputs")
correlations = df.corr()
fig = pyplot.figure(figsize=(8,8))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,14,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```
